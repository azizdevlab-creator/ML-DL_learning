```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1, 100, size = 20)
y = abs(x**2 - 10*x)
y = [i if i != 0 else 1 for i in y]
z = np.log(y)
print(x)
print(y)
print(z)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 8))
axes[0][0].plot(x, y, c = 'teal', marker = 's', markersize = 5, linestyle = '-.', linewidth = 8, label = 'Func chart')
axes[0][0].set_title('First')
axes[0][0].legend()
axes[0][1].scatter(x, z, s = 100, c = np.linspace(1, 100, len(x)), cmap = 'viridis', alpha = 0.8, marker = 'o', edgecolor = 'green')
axes[0][1].set_xlabel('the scatter x')
axes[0][1].grid()
axes[1][0].hist(z, bins = 5, range = (3, 10), color = 'purple', alpha = 0.8, cumulative = False, density = False, edgecolor = 'skyblue')
axes[1][0].set_ylabel('histogram y')
fig.suptitle('The graphs')
fig.supxlabel('X AXES')
plt.tight_layout()
plt.show()
```
