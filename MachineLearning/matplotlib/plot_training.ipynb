```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,3))
axes[0].plot([1, 2, 3], [10, 15, 20], color = 'green')
axes[0].set_title('First')
axes[1].plot([2, 4, 6], [4, 16, 36], color = 'blue')
axes[1].set_title('Second')
plt.tight_layout()
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = [1, 2, 2, 4, 5, 5]
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (12, 5))
axes[0].plot(x, y, c = 'teal')
axes[0].set_title('line chart')
axes[1].scatter(x, y, c = 'b')
axes[1].set_title('Scatter chart')
axes[2].hist(y, color = 'r')
axes[2].set_title('Histogram')

plt.tight_layout()
plt.show()
```

```python
y = np.sin(x)
plt.figure(figsize = (6, 4), edgecolor = 'red', frameon =   True)
plt.plot(x, y,
         color = 'teal',
         linestyle = '--',
         linewidth = 2,
         marker = '*',
         markersize = 10,
         label = 'sinusoida')
plt.legend()
plt.show()
```
