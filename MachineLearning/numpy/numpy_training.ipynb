```python
def test_copy_knowledge():
    """Проверьте свое понимание"""
    a = np.array([[1, 2], [3, 4]])
    
    # Что произойдет?
    b = a[:1, :]  # view или копия?
    c = a.copy()  # какая копия?
    d = a[[0, 1], :]  # view или копия?
    
    b[0, 0] = 999
    print(f"a[0,0] после b: {a[0,0]}")  # 999
    
    c[0, 0] = 111
    print(f"a[0,0] после c: {a[0,0]}")  # 999
    
    d[0, 0] = 222
    print(f"a[0,0] после d: {a[0,0]}")  # 999
```
